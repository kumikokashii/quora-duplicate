```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# Load data

```python
train = pickle.load(open('quora_questions_train_features_v2.pickle', 'rb'))
test = pd.DataFrame()
for i in range(1, 6):
    test = pd.concat([test, pickle.load(open('quora_questions_test_features_v2_part' + str(i) + '.pickle', 'rb'))])
```

# Distriubution of token count

```python
def analysis_token_count(token_count_list):
    plt.hist(token_count_list, bins=range(0, max(token_count_list), 5))
    plt.xlabel('Token Count')
    plt.ylabel('# of Questions')
    plt.title('Distribution of Token Count')
    plt.show()
    
    print('Max:', max(token_count_list), 'tokens')
    for i in [30, 35, 40, 45, 50, 100]:
        print('<= ' + str(i) + ' tokens %:', len(np.where(np.array(token_count_list) <= i)[0]) / len(token_count_list))
    
    plt.hist(token_count_list, bins=range(0, 50))
    plt.xlabel('Token Count')
    plt.ylabel('# of Questions')
    plt.title('Distribution of Token Count')
    plt.show()
```

## Train set questions

```python
analysis_token_count(list(train['token_count_question1']) + list(train['token_count_question2']))
```

## Test set questions

```python
analysis_token_count(list(test['token_count_question1']) + list(test['token_count_question2']))
```
